```python
import pyspark
from pyspark.sql import SparkSession
```

```python
#Createing spark session
spark = SparkSession.builder.appName("NBA").getOrCreate()
```

```python
#Reading Data
dataset = spark.read.csv('Data/nba_logreg.csv', header = True, inferSchema = True)
dataset.show()
```

```python
dataset.dtypes
```

```python
#Selecting multiple columns
dataset.select(['Name', 'PTS']).show()
```

```python
#Summary statistics
dataset.describe().show()
```

```python
type(dataset)
```

```python
#dropping name column
dataset.drop('Name').show()
```

```python
#Creating a new by adding 3 to PTS column
dataset.withColumn('PTS Plus 3', dataset['PTS'] + 3).show()
```

```python
dataset.head(2)
```

```python
#Renaming column
dataset = dataset.withColumnRenamed('PTS', 'PPG')
dataset.show()
```

```python
#renaming column
dataset.withColumnRenamed('TARGET_5Yrs', 'Target')
dataset.show()
```

```python
dataset.count()
```

```python
#droping null values
dataset = dataset.na.drop(how = 'any')
dataset.count()
```
